```python
# prompt: buscar por input campo url  si existe registro mostrar mensaje que se encuentra sitio

import pandas as pd
def buscar_url(url):
  df = pd.DataFrame.from_records(response.json())
  if url in df['url'].values:
    print('El sitio se encuentra registrado como phishing')
  else:
    print('El sitio no se encuentra registrado como phishing')

url = input('Ingrese la URL a buscar: ')
buscar_url(url)
```
